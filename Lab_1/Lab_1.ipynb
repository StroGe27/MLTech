```python
# импорты
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(style="ticks")
```

```python
# 1) Текстовое описание набора данных
# Загрузка данных
data = pd.read_csv('Stock Market Dataset.csv', sep=",")

# добавить описание модели и т.п.
```

```python
# 2) Основные характеристики датасета
data.head()
```

```python
# Размер датасета - 8143 строк, 7 колонок
data.shape
```

```python
total_count = data.shape[0]
print('Всего строк {}'.format(total_count))
```

```python
# Список колонок
data.columns
```

```python
# Список колонок с типами данных
data.dtypes
```

```python
# Проверим наличие пустых значений
# Цикл по колонкам датасета
for col in data.columns:
    # Количество пустых значений - все значения заполнены
    temp_null_count = data[data[col].isnull()].shape[0]
    print('{} - {}'.format(col, temp_null_count))
```

```python
# Основные статистические характеристки набора данных
data.describe()
```

```python
# 3) Визуальное исследование датасета
# Диаграмма рассеяния
fig, ax = plt.subplots(figsize=(10,10))
plt.xticks(np.arange(1, 2000, 200))
plt.yticks(np.arange(1, 2000, 200))
sns.scatterplot(ax=ax, x='Ethereum_Price', y='Bitcoin_Price', data=data)
# можно увидеть что между полями Bitcoin_Price и Ethereum_Price присутствует почти линейная зависимость
```

```python
# Гистограмма
# Позволяет оценить плотность вероятности распределения данных.
fig, ax = plt.subplots(figsize=(10,10))
sns.distplot(data['Crude_oil_Price'])
```

```python
# Jointplot
# Комбинация гистограмм и диаграмм рассеивания.
sns.jointplot(x='Google_Price', y='Apple_Price', data=data)
```

```python
sns.jointplot(x='Google_Price', y='Apple_Price', data=data, kind="hex")
```

```python
sns.jointplot(x='Google_Price', y='Apple_Price', data=data, kind="kde")
```

```python
# "Парные диаграммы"
# Комбинация гистограмм и диаграмм рассеивания для всего набора данных.
# Результат выдал спустя 2 минуты, поэтому уменьшу количество выбираемых данных до 3ех
sns.pairplot(data=data, vars=['Crude_oil_Price', 'Copper_Price', 'Platinum_Price', 'Silver_Price', 'Gold_Price'])
```

```python
# Ящик с усами
# Отображает одномерное распределение вероятности.
sns.boxplot(x=data['Netflix_Price'])
```

```python
# По вертикали
sns.boxplot(y=data['Netflix_Price'])
```

```python
# Violin plot
# Похоже на предыдущую диаграмму, но по краям отображаются распределения плотности
sns.violinplot(x=data['Google_Price'])
```

```python
fig, ax = plt.subplots(2, 1, figsize=(10,10))
sns.violinplot(ax=ax[0], x=data['Google_Price'])
sns.distplot(data['Google_Price'], ax=ax[1])

# Из приведенных графиков видно, что violinplot действительно показывает распределение плотности.
```

```python
# 4) Информация о корреляции признаков
data.corr()
```

```python
data.corr(method='pearson')
```

```python
sns.heatmap(data.corr())

fig, ax = plt.subplots(1, 3, sharex='col', sharey='row', figsize=(15,5))
sns.heatmap(data.corr(method='pearson'), ax=ax[0], annot=True, fmt='.2f')
sns.heatmap(data.corr(method='kendall'), ax=ax[1], annot=True, fmt='.2f')
sns.heatmap(data.corr(method='spearman'), ax=ax[2], annot=True, fmt='.2f')
fig.suptitle('Корреляционные матрицы, построенные различными методами')
ax[0].title.set_text('Pearson')
ax[1].title.set_text('Kendall')
ax[2].title.set_text('Spearman')
```
